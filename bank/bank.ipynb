# Отток клиентов

Проект выполнен с целью прогноза ухода коиентов из банка.
Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет.
Для прогнозирования ухода клиента требуется построить модель машинного обучения с метрикой F1 на тестовой выболрке не менее 0,59
Дополнительно требуется измерить AUC-ROC

## Подготовка данных

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import shuffle
import sklearn.metrics as metrics
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
df = pd.read_csv('D:\\Documents\\ds\\bank\\Churn.csv')
# df = pd.read_csv('/datasets/Churn.csv')

df.info()
print('Количество дубликатов', df.duplicated().sum())
display(df.head())
```

### Вывод  
Предотставленный датасет имеет 10000 строк и 14 столбцов. Пропуски в стобце 'Tenure', дубликатов нет.  
Признаки 'RowNumber', 'CustomerId', 'Surname' не должны оказывать влияние на модели, поэтому должны быть удалены.  
Признаки 'Geography', 'Gender' категориальные, будут преобразованы техникой прямого кодирования.

## Исследование задачи

```python
# Проверка на корреляцию признаков

df2 = df[df['Tenure'].notna()]

plt.figure(figsize=(16, 6))
sns.set(font_scale=1)
matrix = np.triu(df2.corr())
sns.heatmap(df2.corr(), annot=True, cmap= 'coolwarm', mask = matrix)
plt.title('Корреляция признаков')
plt.show()

# Проверка на корреляцию пропущенных значений признака 'Tenure'
df3 = df[df['Tenure'].isna()]
df3 = df.fillna(0)

plt.figure(figsize=(16, 6))
sns.set(font_scale=1)
sns.heatmap(df3.corr(), annot=True, cmap= 'coolwarm', mask = matrix)
plt.title("Корреляция пропущенных значений признака 'Tenure'")
plt.show()
```

```python
# Удаление строк с пропущенными значениями 'Tenure'
df.dropna(subset = ['Tenure'], inplace=True)

# Удаление признаков 'RowNumber', 'CustomerId', 'Surname'
df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)

# Преобразование столбцов 'Geography', 'Gender' техникой прямого кодирования
df = pd.get_dummies(df, columns=['Geography', 'Gender'], drop_first=True)

df.head()
```

```python
# Распределение значений целевого показателя

df['Exited'].value_counts()
```

Были удалены признаки 'RowNumber', 'CustomerId', 'Surname'.  
Признаки 'Geography', 'Gender' преобразованы техникой прямого кодирования.
Корреляции между признаками нет
Корреляции между пропущенными значениями 'Tenure' и другими признакми нет.  
Произвольное заполнение пропущенных данных 'Tenure' может непредсказуемо повлиять на модели обучеия, поэтому строки с пропущенными значениями 'Tenure' удалены.  
Целевой признак 'Exited' бинарный, несбалансированый

## Изучение моделей без учёта дисбаланса классов

```python
# Разбивка данных на выборки
target = df['Exited']
features = df.drop('Exited', axis=1)
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.25,
                                                                              random_state=12345, stratify=df['Exited'])
print(features_train.shape, features_test.shape, target_train.shape, target_test.shape, sep='\n')
```

### Модель "Случайный лес"

```python
parametrs = { 'n_estimators': range (10, 31, 10),
              'max_depth': range (1, 11, 2),
              'min_samples_leaf': range (1, 8),
              'min_samples_split': range (2, 10, 2)
              }

modelF_nb = GridSearchCV(RandomForestClassifier(), parametrs, cv=5, scoring='f1', n_jobs=-1)
modelF_nb.fit(features_train, target_train)
print(modelF_nb.best_estimator_, '%.3f'% modelF_nb.best_score_, sep='\n')
```

### Модель "Дерево решений"

```python
parametrs = { 'criterion': ['gini', 'entropy'],
             'splitter': ['best', 'random'],
             'max_features': ['auto', 'sqrt', 'log2'],
              'max_depth': range(1, 40, 2) }

modelT_nb = GridSearchCV(DecisionTreeClassifier(), parametrs, cv=5, scoring='f1', n_jobs=-1)
modelT_nb.fit(features_train, target_train)
print(modelT_nb.best_estimator_, '%.3f'% modelT_nb.best_score_, sep='\n')
```

### Модель "Логистическая регрессия"

```python
parametrs = {'C': [0.001,0.01,0.1,1,10,100,1000],
                 'solver': ['lbfgs', 'liblinear']}
modelL_nb = GridSearchCV(LogisticRegression(), parametrs, cv=5, scoring='f1', n_jobs=-1)
modelL_nb.fit(features_train, target_train)
print(modelL_nb.best_estimator_, '%.3f'% modelL_nb.best_score_, sep='\n')
```

### Проверка моделей (без учета дисбаланса классов) на валидационной выборке

```python
names = ['Случайный лес', 'Решающее дерево', 'Логистическая регрессия']
models = [modelF_nb, modelT_nb, modelL_nb]
d = dict(zip(names, models))
print('Модель обучения, F1-мера, AUC-ROC', '\n')
for i in d.items():
    pred = i[1].predict(features_test)
    probabilities_test = i[1].predict_proba(features_test)
    probabilities_one_test = probabilities_test[:, 1]
    auc_roc = roc_auc_score(target_test, probabilities_one_test)
    print(f'{i[0]} {f1_score(target_test, pred): .3} {auc_roc: .3}')
```

###  Вывод  
При построении моделей обучения без учета дисбаланса классов наихудшие показатели F1-меры у модели "Логистическая регрессия".
На модели обучения "Случайный лес", "Дерево решений" дисбаланс классов оказывает меньшее влияние.

## Модели с учетом дисбаланса классов

```python
# Устранение дебаланса классов увеличением выборки

features_0 = features_train[target_train == 0]
features_1 = features_train[target_train == 1]
target_0 = target_train[target_train == 0]
target_1 = target_train[target_train == 1]

features_upsampled = pd.concat([features_0] + [features_1] * 4)
target_upsampled = pd.concat([target_0] + [target_1] * 4)
    
features_upsampled, target_upsampled = shuffle(
    features_upsampled, target_upsampled, random_state=12345)
```

### Модель "Случайный лес". Взвешивание классов

```python
parametrs = { 'n_estimators': range(10, 31, 10),
              'max_depth': range(1, 11, 2),
              'min_samples_leaf': range(1, 8),
              'min_samples_split': range(2, 10, 2),
              'class_weight': ['balanced'] }

model_rfr_b = GridSearchCV(RandomForestClassifier(), parametrs, cv=5, scoring='f1', n_jobs=-1)
model_rfr_b.fit(features_train, target_train)
print(model_rfr_b.best_estimator_, '%.2f'% model_rfr_b.best_score_, sep='\n')
```

### Модель "Дерево решений". Взвешивание классов

```python
parametrs = { 'criterion': ['gini', 'entropy'],
             'splitter': ['best', 'random'],
             'max_features': ['auto', 'sqrt', 'log2'],
              'max_depth':range (1, 40, 2), 'class_weight': ['balanced'] }

model_tree_b = GridSearchCV(DecisionTreeClassifier(), parametrs, cv=5, scoring='f1', n_jobs=-1)
model_tree_b.fit(features_train, target_train)
print(model_tree_b.best_estimator_,'%.3f'% model_tree_b.best_score_, sep='\n')
```

### Модель "Логистическая регресиия".  Увеличение выборки

```python
parametrs_lrg = {'C': [0.001,0.01,0.1,1,10,100,1000],
                 'solver': ['lbfgs', 'liblinear'] }

model_lrg = GridSearchCV(LogisticRegression(), parametrs_lrg, cv=5, scoring='f1', n_jobs=-1)
model_lrg.fit(features_upsampled, target_upsampled)
print(model_lrg.best_estimator_, '%.3f'% model_lrg.best_score_, sep='\n')
```

###  Модель "Логистическая регресиия". Взвешивание классов

```python
parametrs = {'C': [0.001,0.01,0.1,1,10,100,1000],
                 'solver': ['lbfgs', 'liblinear'], 'class_weight': ['balanced'] }

model_lrg_b = GridSearchCV(LogisticRegression(), parametrs, cv=5, scoring='f1', n_jobs=-1)
model_lrg_b.fit(features_train, target_train)
print(model_lrg_b.best_estimator_, '%.3f'%  model_lrg_b.best_score_, sep='\n')
```

## Тестирование моделей

### Проверка моделей на валидационной выборке с учетом дисбаланса классов

```python
# Проверка моделей на валидационной выборке

names = ['Случайный лес. Взвешивание классов','Решающее дерево. Взвешивание классов',
         'Логистическая регрессия. Увеличение выборки', 'Логистическая регрессия. Взвешивание классов']
models = [model_rfr_b, model_tree_b, model_lrg, model_lrg_b]
d = dict(zip(names, models))
print('Модель и метод устранения дисбаланса классов, F1-мера, AUC-ROC', '\n')
for i in d.items():
    pred = i[1].predict(features_test)
    probabilities_test = i[1].predict_proba(features_test)
    probabilities_one_valid = probabilities_test[:, 1]
    auc_roc = roc_auc_score(target_test, probabilities_one_test)
    print(f'{i[0]} {f1_score(target_test, pred): .3} {auc_roc: .3}')
```

```python
# Построение ROC-кривой на модели "Случайный лес", показавшей лучшие результаты
probs = model_rfr_b.predict_proba(features_test)
preds = probs[:,1]
fpr, tpr, threshold = metrics.roc_curve(target_test, preds)
roc_auc = metrics.auc(fpr, tpr)


plt.title('ROC-кривая')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.show()
```

### Вывод.  
Наибольшие показатели F1-меры у модели "Случайный лес", наихудшие у модели "Логистическая регрессия"  
Метод "Взвешивание классов" показал лучшие результаты чем метод "Увеличение выборки".

Для прогноза ухода клиента из банка нужно использовать модель "Случайный лес", имеющий самые высокие показатели F1-меры и AUC-ROC.
